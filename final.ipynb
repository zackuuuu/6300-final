```python
import pandas as pd
import glob

dfs = []

# Read all Excel files
for f in glob.glob("/Users/zack/Documents/6300/ttc_bus_delay_data/*.xlsx"):
    df = pd.read_excel(f)
    dfs.append(df)

# Read CSV file and rename columns to match Excel
csv_df = pd.read_csv("/Users/zack/Documents/6300/ttc_bus_delay_data/TTC Bus Delay Data since 2025.csv")
csv_df.rename(columns={
    "_id": "ID",
    "Date": "Report Date",
    "Line": "Route",
    "Station": "Location",
    "Code": "Incident",
    "Bound": "Direction"
}, inplace=True)

dfs.append(csv_df)

# Combine all into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True, sort=False)



# Check first 10 rows
print(combined_df.head(10))
```

```python
combined_df['Report Date'] = combined_df['Report Date'].fillna(combined_df['Date'])
#merge the Date column with Report Date

combined_df = combined_df.drop(columns=['Date'])
#delete the Date column

combined_df['Report Date'] = pd.to_datetime(combined_df['Report Date'], errors='coerce')
# convert the date type to actual date type instead of string type

combined_df = combined_df.sort_values(by='Report Date', ascending=True)
#sort the date into ascending order

combined_df = combined_df.reset_index(drop=True)
# reset the index

combined_df.head(10)
```

```python
df_2016 = combined_df[combined_df['Report Date'].dt.year == 2016]

df_2014 = combined_df[combined_df['Report Date'].dt.year == 2014]

df_2015 = combined_df[combined_df['Report Date'].dt.year == 2015]

print(df_2014.tail(10))

print(df_2014.tail(10))

```

```python
combined_df['Min Gap'] = combined_df['Min Gap'].fillna(combined_df['Gap'])

combined_df = combined_df.drop(columns=['Gap'])

#merge the Gap and Min Gap
```

```python
combined_df['Min Delay'] = combined_df['Min Delay'].fillna(combined_df['Delay'])

combined_df = combined_df.drop(columns=['Delay', 'ID'])

```

```python

# Ensure Report Date is datetime
combined_df['Report Date'] = pd.to_datetime(combined_df['Report Date'], errors='coerce')

# Create Year column
combined_df['Year'] = combined_df['Report Date'].dt.year

# Show first 10 rows for each year
```
